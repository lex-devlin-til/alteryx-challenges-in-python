### 3. Data Transformation

#### Instructions
Financial analysts often need to consolidate data from various sources. This week's challenge focuses on using Alteryx functions in financial planning and analysis (FP&A) projects.

The first dataset   includes sales information for various countries and business segments for the years 2022 and 2023, with the sales figures recorded in each country’s local currency. The second dataset provides the conversion rates from the countries’ currencies to   euros.

Your first task is to create a final table that includes the segment, region, and total sales   in euros for both 2022 and 2023  . (Tip: Remove any columns where the sale price   is null.)

For the second part of the challenge, calculate the year-over-year variation per segment (2022–2023) and identify the segments that experienced negative variations across all three regions: Europe, Asia Pacific (APAC), and the USA.

```python
sales_figures.head()
```

```python
# Create subset of exchange_rates for just Euros
euro_rates = exchange_rates[['Currency', 'EUR']]
```

```python
# Join respective exchange rates onto sales_figures
sales_figures = sales_figures.merge(euro_rates, on=['Currency'], how='left')
```

```python
sales_figures['Sales_in_Euros'] = sales_figures['Sales'] * sales_figures['EUR']
```

```python
sales_figures.head(10)
```

```python
# Deriving year column
sales_figures['Date'] = pd.to_datetime(sales_figures['Date']) # necessary conversion of Date field from object to datetime
sales_figures['Year'] = sales_figures['Date'].dt.year # built-in function that takes the year as an integer
```

### Solution 1:

```python
# Creating subset df for solution 1
sol1 = sales_figures[['Region', 'Segment', 'Year', 'Sales_in_Euros']].copy()
```

```python
# Prepping Year field for pivot
sol1['Year'] = sol1['Year'].astype('string') + ' Total Sales'
```

```python
# Pivot Year and sum sales for each region-segment combo
sol1 = sol1.pivot_table(
    index=['Region', 'Segment'],
    columns='Year',
    values='Sales_in_Euros',
    aggfunc=['sum'])
```

```python
# Solution 1 Complete. Differs from solution table in Alteryx because Euro conversion wasn't done on the data there?
sol1.head()
```

```python
# Pesky "sum" header
sol1.columns = sol1.columns.droplevel(0)
```

```python
sol1.head()
```

```python
# Exporting Solution 1
sol1.to_csv(os.path.join(path, 'solution_1.csv'))
```

```python
sales_figures.drop(columns=['Discount Band', 'COGS', 'Sales', 'Gross Sales'], inplace=True)
```

```python
sales_figures.dtypes
```

```python
sales_figures['2022 Sales'] = sales_figures.loc[sales_figures['Year']==2022, 'Sales_in_Euros'].sum()
sales_figures['2023 Sales'] = sales_figures.loc[sales_figures['Year']==2023, 'Sales_in_Euros'].sum()
sales_figures.head()
```

```python
# Creating yearly sales column for each segment-region
sales_figures['Yearly Sales'] = sales_figures.groupby(['Segment', 'Region', 'Year'])['Sales_in_Euros'].transform(sum)
```

```python
sales_figures.head()
```

```python
sales_figures["Last Year's Sales"] = sales_figures.groupby(['Segment', 'Region', 'Year'])['Yearly Sales'].shift(1)
```

```python
sales_figures.head()
```

```python
sales_sub = sales_figures.groupby(['Segment', 'Region', 'Year']).agg(yearly_sales=('Sales_in_Euros', 'sum')).reset_index()
```

```python
sales_sub.head()
```

```python
sales_sub.sort_values(by=['Segment', 'Region', 'Year'], inplace=True)
sales_sub['lagged_sales'] = sales_sub.groupby(['Segment', 'Region'])['yearly_sales'].shift(1)
```

```python
sales_sub.head()
```

```python
sales_sub['YoY'] = sales_sub['yearly_sales'] - sales_sub['lagged_sales']
sales_sub.head()
```

```python
new_sales = sales_figures.merge(sales_sub, on=['Segment', 'Region', 'Year'], how='left')
```

```python
new_sales.head()
```

```python
new_sales.sort_values(by=['Segment', 'Region', 'Year'], inplace=True)
```
